```python
import cv2
```

```python
path = 'C:/Users/DELL/Documents/TGMT_Image/quangcanh.jpg'
img = cv2.imread(path)
print(img)
```

```python
print(img.shape)
x, y, z = img.shape
print(x)
print(y)
print(z)
```

```python
cv2.imshow('Window', img)
cv2.waitKey(10000)
cv2.destroyWindow('Window')
```

```python
b, g, r = cv2.split(img)
cv2.imshow('Window', img)
cv2.imshow('Red', r)
cv2.imshow('Green', g)
cv2.imshow('Blue', b)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.imshow('Window', img)
cv2.imshow('gray_img', gray_img)
cv2.imshow('img_2', img_2)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
height, width, z = img.shape
quater_h = height/2
quater_w = width/2


green_color = (0,255,0)

img2 = img.copy()


for y in range(int(quater_h)):
    for x in range(int(quater_w)):
        img2[y, x] = green_color
display('Window', img)
display('Window-2', img2)
```
